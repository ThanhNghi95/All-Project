### 2/ Exploring App Genre:


```python
import plotly
plotly.offline.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.express as px
```

```python
num_prime_genre=data['prime_genre'].value_counts()
```

```python
pie=[go.Pie(labels=num_prime_genre.index, values=num_prime_genre.values,
          title="Apple store app genre") ] 
plotly.offline.iplot(pie)
```

### 4/ Feature associate with top 5 App Genre:

##### 4.1/ Price

```python
large_categories=data.groupby(data['prime_genre']).filter(lambda x: len(x)>=248).reset_index()
large_categories['prime_genre'].unique()
```

```python
large_categories['user_rating']
```

```python
trace0 = go.Box(
    # Data for Game
    y=large_categories['price'][large_categories['prime_genre']=='Games'],
    name = 'Game'
)

trace1 = go.Box(
    # Data for Utilities
    y=large_categories['price'][large_categories['prime_genre']=='Utilities'],
    name = 'Utilities'
)

trace2 = go.Box(
    # Data for Entertainment
    y=large_categories['price'][large_categories['prime_genre']=='Entertainment'],
    name = 'Entertainment'
)

trace3 = go.Box(
    # Data for Photo & Video
    y=large_categories['price'][large_categories['prime_genre']=='Photo & Video'],
    name = 'Photo & Video'
)

trace4 = go.Box(
    # Data for Education
    y=large_categories['price'][large_categories['prime_genre']=='Education'],
    name = 'Education'
)
data = [trace0, trace1,trace2,trace3,trace4]
plotly.offline.iplot({'data': data})
```

##### 4.2/ Rating:

```python
trace0 = go.Box(
    # Data for Game
    y=large_categories['user_rating'][large_categories['prime_genre']=='Games'],
    name = 'Game'
)

trace1 = go.Box(
    # Data for Utilities
    y=large_categories['user_rating'][large_categories['prime_genre']=='Utilities'],
    name = 'Utilities'
)

trace2 = go.Box(
    # Data for Entertainment
    y=large_categories['user_rating'][large_categories['prime_genre']=='Entertainment'],
    name = 'Entertainment'
)

trace3 = go.Box(
    # Data for Photo & Video
    y=large_categories['user_rating'][large_categories['prime_genre']=='Photo & Video'],
    name = 'Photo & Video'
)

trace4 = go.Box(
    # Data for Education
    y=large_categories['user_rating'][large_categories['prime_genre']=='Education'],
    name = 'Education'
)
data = [trace0, trace1,trace2,trace3,trace4]
plotly.offline.iplot({'data': data})
```

```python
#### 4.3/ Price and Rating:
sct=[go.Scatter(x=large_categories['price'],
              y=large_categories['user_rating'],
              marker=dict(
                        color=large_categories['prime_genre'].value_counts().index,
                        size=large_categories['prime_genre'].value_counts().values,
             )
               )]               
plotly.offline.iplot(sct)
```
